```python
#Importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,3)
import plotly.express as px
```

```python
#Reading the data
df = pd.read_csv('Levels_Fyi_Salary_Data.csv')
```

```python
#first look at the data
df.head()
```

```python
#Data types for columns

df.dtypes
```

```python
# Dropping unnecesary columns
df.drop('Race', axis=1, inplace=True)
df.drop('Race_Hispanic', axis=1, inplace=True)
df.drop('Race_Asian', axis=1, inplace=True)
df.drop('Race_White', axis=1, inplace=True)
df.drop('Race_Two_Or_More', axis=1, inplace=True)
df.drop('Race_Black', axis=1, inplace=True)
df.drop('otherdetails', axis=1, inplace=True)
df.drop('tag', axis=1, inplace=True)
```

```python
#exploring new dataset
df.head()
```

```python
#checking for null values
df.isnull().sum()
```

```python
#exploring correlation between yearly compensation and stock value
fig = px.scatter(df, x="stockgrantvalue", y="totalyearlycompensation")
fig.show()
```

```python
#correlation
df.corr()
```

```python
#correlation visualization
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix')
plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized
```

```python
#correlation visualization
correlation_matrix = df_numerized.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix')
plt.show()
```

```python
correlation_mat = df_numerized.corr()

corr_par = correlation_mat.unstack()

corr_par
```

```python
sorted_pair = corr_par.sort_values()
sorted_pair
```

```python
high_corr = sorted_pair[(sorted_pair)>0.5]
high_corr
```
